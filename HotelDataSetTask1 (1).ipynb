```python
import pandas as pd

df = pd.read_csv('first inten project.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x='booking status', y='lead time', data=df)
plt.title('Booking Status vs. Lead Time')
plt.show()
```

```python
# Visualize the relationship between 'average price' and 'booking status'
plt.figure(figsize=(10, 6))
sns.boxplot(x='booking status', y='average price ', data=df)
plt.title('Booking Status vs. Average Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.countplot(x='market segment type', hue='booking status', data=df)
plt.title('Booking Status by Market Segment Type')
plt.xlabel('Market Segment Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
